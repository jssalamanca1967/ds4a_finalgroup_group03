```python
#import necessary libraries
import numpy as np 
import pandas as pd
import plotly
#import plotly.plotly as py
import plotly.offline as offline
import plotly.graph_objs as go

import folium  #needed for interactive map
from folium.plugins import HeatMap

import branca

import geopandas
from shapely.geometry import Polygon
from shapely.geometry import Point
from geopandas import GeoDataFrame
```

```python
file = 'departamentos_colombia.geojson'
#file = 'https://raw.githubusercontent.com/jackparmer/mapbox-counties/master/2003/0-2.geojson'


colombia = geopandas.read_file(file)
colombia.head()
```

```python
colombia.dropna(inplace = True)
colombia.head()
```

```python
from geojson import Point, Feature, FeatureCollection, dump

with open('colombia_limpio.geojson', 'w') as f:
   dump(colombia, f)
```

```python
# interactive visualization for beat-specific crime rate in 2017

m_trips = folium.Map(location=[4.570868, -74.2973328],
                        zoom_start=3,
                        tiles="OpenStreetMap")

colormap = branca.colormap.LinearColormap(
    colors=['white','yellow','orange','red','darkred'],
    #index=beat_cn['count'].quantile([0.2,0.4,0.6,0.8]),b
    vmin=colombia.n_universidades.min(),
    vmax=colombia.n_universidades.max()
)

style_function = lambda x: {
    'fillColor': colormap(x['properties']['n_universidades']),
    'color': 'black',
    'weight':2,
    'fillOpacity':0.5
}

stategeo = folium.GeoJson(
    colombia.to_json(),
    name='Chicago beats',
    style_function=style_function,
    tooltip=folium.GeoJsonTooltip(
        fields=['departamento', 'n_universidades'],
        aliases=['Departamento', 'Num Universidades'], 
        localize=True
    )
).add_to(m_trips)

colormap.add_to(m_trips)

m_trips.save("mapa_colombia.html")
```
